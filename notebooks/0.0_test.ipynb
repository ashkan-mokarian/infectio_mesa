```python
import sys

from infectio_mesa.model import BasicModel
import numpy as np
import matplotlib.pyplot as plt
from matplotlib_inline import backend_inline
backend_inline.set_matplotlib_formats('svg')
%matplotlib inline
```

```python
def draw(model):
    x_vals = []
    y_vals = []
    for cell in model.schedule.agents:
        x, y = cell.pos
        x_vals.append(x)
        y_vals.append(y)
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111)
    ax.scatter(x_vals, y_vals)
```

```python
model = BasicModel(100, 100, 100)
```

```python
for _ in range(10):
    model.step()
```

```python
draw(model)
```
